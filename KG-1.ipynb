```python
from indicnlp.tokenize import indic_tokenize
import re

# Telugu text to tokenize
telugu_text = sentences
sentence_tokens = {}

# Regular expression pattern to match Telugu characters and whitespace
telugu_pattern = re.compile(r'[\u0C00-\u0C7F\s]+')

# Tokenize the Telugu text
for sent_index in range(len(sentences)):
    tokens = indic_tokenize.trivial_tokenize(sentences[sent_index])
    print("Token:",tokens)
    normalized_tokens = []
    for token in tokens:
        # Remove characters that are not Telugu characters or whitespace
        normalized_token = re.sub(r'[^\u0C00-\u0C7F\s]', '', token)
        if normalized_token!='':
            normalized_tokens.append(normalized_token)
    sentence_tokens[sent_index] = normalized_tokens

    # Print the tokens
    print("Normalized_token:", sentence_tokens[sent_index])
```

```python
dependency_relations = []

for doc in sentence_info:
    for sent in doc.sentences:
        added_relation = False
        for word in sent.words:
            if word.upos not in 'PUNCT' and 0 < word.head <= len(sent.words):
                relation = sent.words[word.head-1].text
                if relation not in dependency_relations and not added_relation:
                    dependency_relations.append(relation)
                    added_relation = True

```

```python
ht = []
tt = []
desired_entities = 50

for doc in sentence_info:
    for sent in doc.sentences:
        head_found = False
        tail_found = False
        
        for word in sent.words:
            if word.deprel not in 'PUNCT':
                # Search for head entity
                if (word.deprel == 'nsubj' or word.deprel == 'list') and (word.xpos == 'NOUN' or word.xpos == 'PROPN') and not head_found:
                    ht.append(word.text)
                    head_found = True
                # Search for tail entity
                if (word.deprel == 'obj' or word.deprel == 'obl') and (word.xpos == 'NOUN' or word.xpos == 'PROPN' or word.xpos == 'VERB') and not tail_found:
                    tt.append(word.text)
                    tail_found = True
                
            if head_found and tail_found:
                break  # Break if both head and tail entities found in the current sentence
        
        # Check if entities found for the current sentence, if not, append empty strings
        if not head_found:
            ht.append('NA')
        if not tail_found:
            tt.append('NA')
        
        if len(ht) >= desired_entities and len(tt) >= desired_entities:
            break  # Break if desired number of entities reached
```

```python
def map_entities(head_entities, relations, tail_entities):
    entity_mapping = {}
    relation_mapping = {}
    tail_mapping = {}
    
    # Mapping head entities
    for index, entity in enumerate(head_entities):
        numeric_value = index + 1
        entity_mapping.setdefault(entity, []).append(numeric_value)
    
    # Mapping relations
    for index, relation in enumerate(relations):
        numeric_value = index + 51
        relation_mapping.setdefault(relation, []).append(numeric_value)
    
    # Mapping tail entities
    for index, entity in enumerate(tail_entities):
        numeric_value = index + 101
        tail_mapping.setdefault(entity, []).append(numeric_value)
    
    return entity_mapping, relation_mapping, tail_mapping
# Mapping
head_mapping, relation_mapping, tail_mapping = map_entities(ht, dependency_relations , tt)

# Print the mappings
print("Head Entities Mapping:")
for entity, numeric_value in head_mapping.items():
    print(f"{entity}: {numeric_value}")

print("\nRelations Mapping:")
for relation, numeric_value in relation_mapping.items():
    print(f"{relation}: {numeric_value}")

print("\nTail Entities Mapping:")
for entity, numeric_value in tail_mapping.items():
    print(f"{entity}: {numeric_value}")
```

```python
from gensim.models import Word2Vec
# Function to train Word2Vec model and generate embeddings
def generate_embeddings(sentence_list):
    model = Word2Vec(sentences=[sentence_list], vector_size=64, window=5, min_count=1)
    model_vocabulary = set(model.wv.index_to_key)
    word_embeddings = {word: model.wv[word].tolist() for word in model_vocabulary}  
    # Fill embeddings for words not in vocabulary with zeros
    for word in sentence_list:
        if word not in model_vocabulary:
            word_embeddings[word] = [0] * model.vector_size
    
    return word_embeddings

# Generate embeddings for each list
embeddings_head = generate_embeddings(ht)
embeddings_rel = generate_embeddings(dependency_relations)
embeddings_tail = generate_embeddings(tt)

# Print out words not present in the model's vocabulary for each list
print("Words not present in the model's vocabulary for list1:", [word for word in ht if word not in embeddings_head])
print("Words not present in the model's vocabulary for list2:", [word for word in dependency_relations if word not in embeddings_rel])
print("Words not present in the model's vocabulary for list3:", [word for word in tt if word not in embeddings_tail])

# Print out embeddings for each list
print("\nEmbeddings for list1:")
for word, embedding in embeddings_head.items():
    print(f"{word}: {embedding}")

print("\nEmbeddings for list2:")
for word, embedding in embeddings_rel.items():
    print(f"{word}: {embedding}")

print("\nEmbeddings for list3:")
for word, embedding in embeddings_tail.items():
    print(f"{word}: {embedding}")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
def link_prediction(node1, node2, threshold, word_embeddings):
    if node1 in word_embeddings and node2 in word_embeddings:
        # Get embeddings of nodes
        node1_embedding = word_embeddings[node1]
        node2_embedding = word_embeddings[node2]
        
        # Calculate similarity between embeddings
        similarity = cosine_similarity([node1_embedding], [node2_embedding])[0][0]
        
        # Predict link based on similarity and threshold
        if similarity >= threshold:
            return True
        else:
            return False
    else:
        return False  # Return False if embeddings not found for nodes

threshold = 0.2 # Adjust threshold as needed
while True:
    # Take input from the user
    node1 = input("Enter word1: ")
    node2 = input("Enter word2: ")

    is_linked = link_prediction(node1, node2, threshold, word_embeddings)
    print(f"Are '{node1}' and '{node2}' linked? {is_linked}")
    
    # Ask user if they want to continue
    continue_checking = input("Do you want to continue? (y/n): ")
    if continue_checking.lower() != 'y':
        break
is_linked = link_prediction(node1, node2, threshold, word_embeddings)
print(f"Are '{node1}' and '{node2}' linked? {is_linked}")
```

```python
import numpy as np

class TransE:
    def __init__(self, embeddings, num_relations, learning_rate=0.01, margin=0.5):
        self.head_embeddings, self.relation_embeddings, self.tail_embeddings = embeddings
        self.num_entities = min(len(self.head_embeddings), len(self.tail_embeddings))
        self.embedding_size = len(list(self.head_embeddings[0].values()))  # Assuming embeddings are of same size
        self.num_relations = num_relations
        self.learning_rate = learning_rate
        self.margin = margin
        
    def train(self, triples, epochs=1000):
        for epoch in range(epochs):
            loss = 0
            for triplet in triples:
                if "NA" in triplet:
                    continue  # Skip triplets with missing entities
                
                head, relation, tail = triplet
                
                if head not in self.head_embeddings or tail not in self.tail_embeddings:
                    continue # Skip triplets with out-of-bound entities
                
                # Look up embeddings
                head_embedding = self.head_embeddings[head]
                tail_embedding = self.tail_embeddings[tail]
                relation_embedding = self.relation_embeddings[relation]
                
                # Calculate score
                score = np.linalg.norm(head_embedding + relation_embedding - tail_embedding, ord=1)
                
                # Calculate loss
                current_loss = max(0, self.margin + score - np.linalg.norm(head_embedding - relation_embedding + tail_embedding, ord=1))
                loss += current_loss
                
                # Update embeddings
                if current_loss > 0:
                    # Calculate gradients
                    gradient_head = -(tail_embedding - head_embedding - relation_embedding)
                    gradient_tail = (head_embedding + relation_embedding - tail_embedding)
                    gradient_relation = (tail_embedding - head_embedding - relation_embedding)
                    
                    # Update embeddings
                    self.head_embeddings[head] -= self.learning_rate * gradient_head
                    self.tail_embeddings[tail] -= self.learning_rate * gradient_tail
                    self.relation_embeddings[relation] -= self.learning_rate * gradient_relation

            print("Epoch {}: Loss = {}".format(epoch+1, loss))
            
            # Moved print statements outside of the loop
# Example usage
# Assuming you have three lists of embeddings called "head_embedding", "relation_embedding", and "tail_embedding"
# Where each inner list contains dictionaries mapping entities to embedding vectors
# Dummy triples data with NA (replace with your actual data)
num_relations = 50
model = TransE(word_embedding, num_relations)
model.train(triplets)
```
