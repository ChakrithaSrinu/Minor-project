```python
from googletrans import Translator
from langdetect import detect

def translate_telugu_to_english(telugu_text):
    translator = Translator()
    english_text = translator.translate(telugu_text, src='te', dest='en').text
    return english_text

def process_text(input_text):
    detected_language = detect(input_text)

    if detected_language == "te":
        return translate_telugu_to_english(input_text)
    else:
        return input_text

def process_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        sentences = file.read().splitlines()
        translated_sentences = []

        for sentence in sentences:
            translated_sentence = process_text(sentence)
            translated_sentences.append(translated_sentence.lower())

    return translated_sentences

# Provide the path to your text file
file_path = './Telugu.txt'

translated_sentences = process_file(file_path)

for translated_sentence in translated_sentences:
    print(translated_sentence)
```

```python
import spacy

def get_entities(sentences):
    entity_pairs = []  # List to store entity pairs for each sentence
    
    nlp = spacy.load("en_core_web_sm")
    
    for sent in sentences:
        ent1 = ""
        ent2 = ""
        prv_tok_dep = ""   
        prv_tok_text = ""   
        prefix = ""
        modifier = ""
        
        for tok in nlp(sent):
            if tok.dep_ != "punct":
                if tok.dep_ == "compound":
                    prefix = tok.text
                    if prv_tok_dep == "compound":
                        prefix = prv_tok_text + " " + tok.text
    
                if tok.dep_.endswith("mod") == True:
                    modifier = tok.text
                    if prv_tok_dep == "compound":
                        modifier = prv_tok_text + " " + tok.text
    
                if tok.dep_.find("subj") == True:
                    ent1 = modifier + " " + prefix + " " + tok.text
                    prefix = ""
                    modifier = ""
                    prv_tok_dep = ""
                    prv_tok_text = ""
    
                if tok.dep_.find("obj") == True:
                    ent2 = modifier + " " + prefix + " " + tok.text
    
                prv_tok_dep = tok.dep_
                prv_tok_text = tok.text
        
        entity_pairs.append([ent1.strip(), ent2.strip()])
    
    return entity_pairs

# Example usage with a list of sentences
entities_list = get_entities(translated_sentences)

for entities in entities_list:
    print(entities[0], "->", entities[1])
```

```python
import spacy
from spacy.matcher import Matcher

def get_relation(sent):
    nlp = spacy.load("en_core_web_sm")  # Load the English language model
    doc = nlp(sent)
    
    # Matcher class object
    matcher = Matcher(nlp.vocab)
    
    # Define the pattern
    pattern = [{'DEP':'ROOT'},
               {'DEP':'prep','OP':"?"},
               {'DEP':'agent','OP':"?"},
               {'POS':'ADJ','OP':"?"}]
    
    matcher.add("matching_1", [pattern])
    
    matches = matcher(doc)
    k = len(matches) - 1
    
    span = doc[matches[k][1]:matches[k][2]]
    
    return span

rel = []
for sentence in translated_sentences:
    r = get_relation(sentence)
    print(r)
    rel.append(r.text)
```

```python
from googletrans import Translator
from langdetect import detect

def en_to_te_list(text_list):
    translator = Translator()
    telugu_list = []

    for sublist in text_list:
        s, o, r = sublist
        s_telugu = translator.translate(s, src='en', dest='te').text
        o_telugu = translator.translate(o, src='en', dest='te').text
        r_telugu = translator.translate(r, src='en', dest='te').text
        telugu_list.append([s_telugu, o_telugu, r_telugu])

    return telugu_list


tel_triplets = en_to_te_list(v)

for a in tel_triplets:
    s, o, r = a
    print(s, "-", o, "-", r)
```

```python
telugu_dicts = []

for sublist in tel_triplets:
    telugu_dict = {'subject': sublist[0], 'object': sublist[1], 'relation': sublist[2]}
    telugu_dicts.append(telugu_dict)

print(telugu_dicts)
print(len(telugu_dicts))
```

```python
from gensim.models import Word2Vec

# Train a Word2Vec model
model = Word2Vec(sentences=tel_triplets, vector_size=300, window=5, min_count=1)

# Get word embeddings for each word in each sublist
word_embeddings = {word: model.wv[word].tolist() for triplet in tel_triplets for word in triplet}

# Display word embeddings
for word, embedding in word_embeddings.items():
    print(f"{word}: {embedding}")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

triplets = telugu_dicts.copy()
embedding_vectors = {}
for i,j in word_embeddings.items():
    embedding_vectors[i] = np.array(j)

def find_missing_word(word1, word2):
    # Calculate the average embedding of the input words
    avg_embedding = (embedding_vectors[word1] + embedding_vectors[word2]) / 2
   
    # Calculate cosine similarity between the average embedding and the third word in the triplet
    similarities = {}
    for triplet in triplets:
        if word1 in triplet.values() and word2 in triplet.values():
            for word in triplet.values():
                if word != word1 and word != word2:
                    similarity = cosine_similarity(avg_embedding.reshape(1, -1), embedding_vectors[word].reshape(1, -1))[0][0]
                    similarities[word] = similarity
   
    # Find the word with the highest cosine similarity
    missing_word = max(similarities, key=similarities.get)
    return missing_word, embedding_vectors[missing_word]

x = True
while x:
    print("\nInput any two Telugu words of your choice\n")
    word1 = input("Word-1 : ")
    word2 = input("Word-2 : ")
   
    try:
        if (word1 != word2):
            missing_word, embedding_vector = find_missing_word(word1, word2)
            print("Missing word in the triplet is ", missing_word)
            print("Embedding vector : ", embedding_vector)
        else:
            print("\nTwo inputted words are same..")
    except ValueError:
        print("\nNo triplet found containing both inputted words!!")

    c = input("\nWant to try another (y/n) : ").lower()
    while c not in ["y", "n"]:
        print("Invalid input! Please enter 'y' or 'n'.")
        c = input("\nWant to try another (y/n) : ").lower()

    if c == "y":
        x = True
    else:
        x = False
```
