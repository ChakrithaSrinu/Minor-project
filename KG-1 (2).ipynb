```python
from indicnlp.tokenize import indic_tokenize
import re

# Telugu text to tokenize
telugu_text = sentences
sentence_tokens = {}

# Regular expression pattern to match Telugu characters and whitespace
telugu_pattern = re.compile(r'[\u0C00-\u0C7F\s]+')

# Tokenize the Telugu text
for sent_index in range(len(sentences)):
    tokens = indic_tokenize.trivial_tokenize(sentences[sent_index])
    print("Token:",tokens)
    normalized_tokens = []
    for token in tokens:
        # Remove characters that are not Telugu characters or whitespace
        normalized_token = re.sub(r'[^\u0C00-\u0C7F\s]', '', token)
        if normalized_token!='':
            normalized_tokens.append(normalized_token)
    sentence_tokens[sent_index] = normalized_tokens

    # Print the tokens
    print("Normalized_token:", sentence_tokens[sent_index])
```
